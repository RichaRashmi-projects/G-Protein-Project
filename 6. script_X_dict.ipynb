funtion to get all the possible 7mers with one X in it.

```
def list_of_7mer_X(sevenmer):
    x_data = []
    for r1 in range(7):
        x=list(sevenmer)
        x[r1]="X"
        x= ''.join(x)
        x_data.append(x)
    return x_data
```

```
sheet_path= "/content/drive/My Drive/Colab Notebooks/GProtein_Project/7mer_sheet.csv"
data_sheet = pd.read_csv(sheet_path)
```

Insert X at all possible positions in G5-Box from 'SA' with 1 mismatch G-Domains

```
for x in range(20):
    protein_family = data_sheet.loc[x , "Protein_Class"]
    print (protein_family)
    if (family==protein_family):
        print (protein_family)
        inpath_SA_mm= data_sheet.loc[x , "Inpath_SA_mm"]
        outpath1_SA_mm= data_sheet.loc[x , "Outpath_X_SA_mm"]
        outpath2_SA_mm= data_sheet.loc[x , "Outpath_X_SA_mm_count"]

        total= data_sheet.loc[x , "Count"]

        data = pd.read_csv(inpath_SA_mm)

        unique_7mers= data['G5-box'].unique()
        temp= data

        id_set = {}

        for j in range(temp.shape[0]):
            seq= temp.loc[j,"G5-box"]
            ID= temp.loc[j,"ProteinID"]
            
            x_data = list_of_7mer_X(seq)
            for x in x_data:
                if(x not in id_set):
                    id_set[x] = set()
                    id_set[x].add(ID)
                else:
                    id_set[x].add(ID)
    
```

```
id_set.items()
```

Write the results to the drive.

```
with open(outpath1_SA_mm, 'w') as f:   
    [f.write('{0},{1}\n'.format(key, value)) for key, value in id_set.items()]

with open(outpath2_SA_mm, 'w') as f:
    [f.write('{0},{1}\n'.format(key, [len(value),round((100 * len(value) / total),2)])) for key, value in id_set.items()]
```
