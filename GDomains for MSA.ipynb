```
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```
inpath_info = "/content/drive/My Drive/Colab Notebooks/GProtein_Project/ProteinIDs_Before.csv"
info = pd.read_csv(inpath_info)
```

```
info.head()
```

```
list_seq = list()
list_name = list()
```

```
for i in range(info.shape[0]):
    
    family = info.loc[i , "G protein class"]
    ID = info.loc[i, "Protein ID"]
    start = int(info.loc[i , "Before G1 start site"])+1
    end = int(info.loc[i , "Before G5 start site"])+8
    
    inpath_family = "/content/drive/My Drive/Colab Notebooks/GProtein_Project/"+family+"/"+family+".xlsx"
    df1 = pd.read_excel(inpath_family)
    #df2 = df1.set_index("Entry", drop = False)
    
    for j in range(df1.shape[0]):
        if(df1.loc[j , "Entry"] == ID):
            sequence = df1.loc[j , "Sequence"]
    
    list_seq.insert(i, sequence[start:end]) 
    #list_seq.insert(i, sequence)
    list_name.insert(i, (family+"_"+ID+"_"+str(i)))    
    #list_seq[i] = sequence[start:end]
    #list_name[i] = family
```

```
list_name
```

```
ofile = open("/content/drive/My Drive/Colab Notebooks/GProtein_Project/Gdomain_Before.fasta", "w")
#ofile = open("H:/uniprot/Gdomain_Before_fulllength.fasta", "w")
```

```
for i in range(len(list_seq)):
    ofile.write(">" + list_name[i] + "\n" +list_seq[i] + "\n")

ofile.close()
```
